# Цель работы:

Научиться использовать систему контроля версий `GIT` для управления локальным и удалённым репозиторим, освоить создание веток, слияние изменений, а также размещение сайтов на GitHub.

### **Ход работы:**

1. **Установка `GIT`:**

   * Скачал и установил `Git-Bash `с официального сайта [git-scm.com](https://git-scm.com/install/windows).

2. **Создание локального репозитория:**

   * Для начала переходим в католог где создадим папку. Создаём папку для проекта и инициализируем `GIT`:
   
     ```bash
     cd "C:\Users\user\Desktop\labs\git"
     mkdir homework_IIT
     cd homework_IIT
     git init
     ```


3. **Создание файла отчёта и фиксация изменений:**

   * Создал файл отчета `Системы контроля версии_Чекурин_Кирилл.ipynb` и описал шаги выполнения практической работы.
   * Проиндексировал файл и зафиксировал изменения:

     ```bash
     git add Системы_контроля_версии_Чекурин_Кирилл.ipynb
     git commit -m "Создан файл отчета с описанием проделанных шагов"
     ```

4. **Создание ветки и работа с ней:**

   * Создал новую ветку для внесения дополнительных описаний:

     ```bash
     git branch report_branch
     git checkout report_branch
     ```
   * Добавил подробное описание шагов в отчет и зафиксировал изменения:

     ```bash
     git commit -am "Добавлены подробные описания шагов"
     ```

5. **Слияние изменений в основную ветку:**

   * Переключился на основную ветку и слил изменения:

     ```bash
     git checkout master
     git merge report_branch
     ```

6. **Создание репозитория на `GitHub` и загрузка проекта:**

   * Вошёл в `GitHub` и создал репозиторий `homework_IIT`.
   * Связал локальный репозиторий с удалённым и отправил изменения:

     ```bash
     git remote add origin https://github.com/pocketgodru/homework-IIT.git
     git push -u origin master
     ```

7. **Размещение сайтов на `GitHub`:**

   * Через веб-браузер загрузил отдельно ранее созданный сайт
   * Разметил сайт на GitHub Pages.


Github репозиторий созданный мной: [pocketgodru/homework-IIT](https://github.com/pocketgodru/homework-IIT)

Сайт размещенный в Github-Pages: [homework-IIT](https://pocketgodru.github.io/homework-IIT/)

### **Вывод:**

GIT — это очень полезная и удобная система контроля версий. Она позволяет хранить историю изменений, работать с разными ветками и безопасно вносить правки в проект. Благодаря GIT можно легко совместно работать над проектами и загружать в удалённые репозитории, что делает управление проектами более организованным и эффективным.



### **Трудности:**

* Путаница названий главной ветки не `main`, а `master`
