# Цель работы:

Научиться использовать систему контроля версий `GIT` для управления локальным и удалённым репозиторим, освоить создание веток, слияние изменений, а также размещение статических и динамических сайтов на GitHub.

### **Ход работы:**

1. **Установка GIT:**

   * Скачал и установил Git-Bash с официального сайта [git-scm.com](https://git-scm.com/install/windows).

2. **Создание локального репозитория:**

   * Для начала переходим в католог где создадим папку. Создаём папку для проекта и инициализируем GIT:
   
     ```bash
     cd "C:\Users\user\Desktop\labs\git"
     mkdir homework_IIT
     cd homework_IIT
     git init
     ```


3. **Создание файла отчёта и фиксация изменений:**

   * Создал файл отчета `Системы контроля версии_Чекурин_Кирилл.ipynb` и описал в нём все шаги выполнения практической работы.
   * Добавил файл в индекс и зафиксировал изменения:

     ```bash
     git add Системы контроля версии_Чекурин_Кирилл.ipynb
     git commit -m "Создан файл отчета с описанием проделанных шагов"
     ```

4. **Создание ветки и работа с ней:**

   * Создал новую ветку для внесения дополнительных описаний:

     ```bash
     git branch report_branch
     git checkout report_branch
     ```
   * Добавил подробное описание шагов в отчет и зафиксировал изменения:

     ```bash
     git commit -am "Добавлены подробные описания шагов"
     ```

5. **Слияние изменений в основную ветку:**

   * Переключился на основную ветку и слил изменения:

     ```bash
     git checkout main
     git merge report_branch
     ```

6. **Создание репозитория на GitHub и загрузка проекта:**

   * Зарегистрировался на GitHub и создал репозиторий `homework_IIT`.
   * Связал локальный репозиторий с удалённым и отправил изменения:

     ```bash
     git remote add origin https://github.com/pocketgodru/homework-IIT.git
     git push -u origin main
     ```

7. **Размещение сайтов на GitHub:**

   * Разместил ранее созданный статический сайт в репозитории.
   * (Дополнительно) Разметил динамический сайт на GitHub Pages.


---

### **Вывод:**

В ходе практической работы я освоил базовые команды GIT: создание локального репозитория, добавление и фиксацию изменений, работу с ветками и слияние изменений. Также научился загружать проекты на GitHub и размещать статические и динамические сайты. Это позволяет вести контроль версий и работать с удалёнными репозиториями, что является важным навыком для совместной разработки.
